```python
import pyautogui
import time

pyautogui.alert("O cógido começará a ser executado, retire suas mãos do teclado e do mouse")
pyautogui.PAUSE = 0.5
# abrir o google drive
pyautogui.press('winleft')
pyautogui.write('chrome')
pyautogui.press('enter')
time.sleep(1)
pyautogui.write('https://drive.google.com/drive/my-drive')
pyautogui.press('enter')

# entrar na área de trabalho
pyautogui.hotkey('winleft','d')

# cliquei no arquivo que eu quero fazer backup e arrastei ele
pyautogui.moveTo(1239, 507)
time.sleep(2)
pyautogui.mouseDown()
time.sleep(2)
pyautogui.moveTo(664, 449)
time.sleep(2)

# enquanto eu tô arrastando eu vou mudar para o drive
pyautogui.hotkey('alt','tab')
time.sleep(2)
# larguei ele no google drive
pyautogui.mouseUp()

# esperar 5 segundos
time.sleep(10)

#alerta de fim do processo
pyautogui.alert("Processo finalizado, pode usar seu computador de novo")

#shift enter e atalho para executar A celula pelo teclado
#A posição do mouse varia de tela pra tela
```
